```python
skyg = enmap.fullsky_geometry(10*utils.arcmin)
sky = enmap.zeros(skyg[0],skyg[1])
```

```python
list = np.sort(gl.glob(release_path+"/masks/footprints/act_dr4_*_apo_w*.fits"))
list = np.delete(list,7)
list = np.delete(list,7)

for i, file in enumerate(list):
    print(file)   
    foot  = enmap.read_map(file)
    foot[foot<1.] = 0
    down = utils.nint(sky.wcs.wcs.cdelt[1]/foot.wcs.wcs.cdelt[1])
    foot = enmap.downgrade(foot,down)
    foot = enmap.project(foot, sky.shape, sky.wcs, order=0)
    
    if i==0: 
        layer = enplot.plot(foot, grid=False, layers=True, contour_width=2, no_image=True, contours="0.5")
    else:
        layer += enplot.plot(foot, grid=False, layers=True, contour_width=2, no_image=True, contours="0.5")
```

```python
p_sky = enplot.merge_plots(layer)
enplot.show(p_sky)
enplot.write('all_bound.png', p_sky)
```
